```python
import pandas as pd

df = pd.read_csv("/content/linkedin_jobs_with_extracted_skills_per_job4 (2).csv")
```

```python
skill_replacements = {
    "c + +": "c++",
    "ci/cd": "CI/CD",
    "test cases": "test case design",
    "trajectory": "test automation",
    "mecha": "automation testing",
    "algorithm development": "algorithm design",
    "performance tuning": "performance testing",
    "code analysis": "static code analysis",
    "defect analysis": "defect management"
}
```

```python
def replace_skills(skill_string):

    if pd.isna(skill_string):
        return skill_string

    skills = skill_string.split(",")

    cleaned_skills = []

    for skill in skills:
        skill = skill.strip().lower()

        if skill in skill_replacements:
            skill = skill_replacements[skill]

        cleaned_skills.append(skill)

    return ", ".join(cleaned_skills)
```

```python
mask = df["final_category"] == "QA & Testing"

df.loc[mask, "extracted_skills"] = df.loc[mask, "extracted_skills"].apply(replace_skills)
```

```python



df.loc[mask, "extracted_skills"] = df.loc[mask, "extracted_skills"].str.replace("hybrid", "hybrid testing")
df.loc[mask, "extracted_skills"] = df.loc[mask, "extracted_skills"].str.replace("localization", "localization testing")

```

```python
df.to_csv("linkedin_jobs_cleaned_skills.csv", index=False)
```
